**EDA - EXPLORATORY DATA ANALYSIS FOR SCENARIO 3 (Money at a fixed rate for a business for a construction loan)**

```python
# Load the dataset
Data_3 = pd.read_csv('Construction Loan Dataset.csv')
Data_3.head()
```

```python
Data_3.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.distplot(Data_3['TTLCONS'], kde=True, color='skyblue')
plt.title('Distribution of Total Construction Spending')
plt.xlabel('Total Construction Spending')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.boxplot(Data_3["TTLCONS"], vert=False)
plt.xlabel("Millions of USD")
plt.title("Distribution of Total Construction Spending");
```

```python
Data_3[['TTLCONS', 'PCU32733273', 'FEDFUNDS', 'Real GDP']].corr()
```

```python
sns.regplot(x="PCU32733273", y="TTLCONS", data=Data_3)
```

```python
Data_3.plot(kind='scatter', x='PCU32733273', y='TTLCONS', figsize=(10, 6), color='green')

plt.title('Relationship between Total Construction Spending and Material Cost')
plt.xlabel('Material Cost')
plt.ylabel('Total Construction Spending')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=Data_3, x='FEDFUNDS', y='TTLCONS', color='purple')
plt.title('Relationship between Total Construction Spending and Material Cost')
plt.xlabel('Interest Rate')
plt.ylabel('Total Construction Spending')

plt.show()
```

```python
fig, ax = plt.subplots()

x = Data_3['Date']
y1 = Data_3['TTLCONS']
y2 = Data_3['Real GDP']
# Plot first data series
ax.plot(x, y1, label='Total Construction Spending', marker='o', linestyle='-')

# Plot second data series
ax.plot(x, y2, label='Real GDP', marker='s', linestyle='--')

# Add labels and title
ax.set_xlabel('Date')
ax.set_ylabel('Value')
ax.set_title('Line Chart')

# Add a legend
ax.legend()

# Display the plot
plt.show()
```

```python
fig, ax = plt.subplots()
# Plot second data series
ax.plot(x, y2, label='Real GDP', marker='s', linestyle='--')

# Add labels and title
ax.set_xlabel('Date')
ax.set_ylabel('Value')
ax.set_title('Line Chart')

# Add a legend
ax.legend()

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=Data_3, x='Real GDP', y='TTLCONS', color='black')
plt.title('Relationship between Total Construction Spending and Economy')
plt.xlabel('Real GDP')
plt.ylabel('Total Construction Spending')

plt.show()
```

**EDA - EXPLORATORY DATA ANALYSIS FOR SCENARIO 4 (Publicly traded equity)**

```python
# Load the dataset
Data_4 = pd.read_csv('Publicly Traded Equity Dataset.csv')
Data_4.head()
```

```python
Data_4.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.distplot(Data_4['DRBLACBS'], kde=True, color='skyblue')
plt.title('Distribution of Delinquency Rate on Business Loans')
plt.xlabel('Delinquency Rate on Business Loans')
plt.ylabel('Frequency')
plt.show()
```

```python
Data_4[['FEDFUNDS', 'CPIAUCSL', 'DRBLACBS', 'SP500']].corr()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=Data_4, x='DRBLACBS', y='FEDFUNDS', color='black')
plt.title('Relationship between Delinquency Rate and Interest Rate')
plt.xlabel('Interest Rate')
plt.ylabel('Delinquency Rate')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=Data_4, x='DRBLACBS', y='CPIAUCSL', color='Green')
plt.title('Relationship between Delinquency Rate and CPI')
plt.xlabel('CPI')
plt.ylabel('Delinquency Rate')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=Data_4, x='DRBLACBS', y='SP500', color='Yellow')
plt.title('Relationship between Delinquency Rate and Stock market')
plt.xlabel('SP500')
plt.ylabel('Delinquency Rate')

plt.show()
```
